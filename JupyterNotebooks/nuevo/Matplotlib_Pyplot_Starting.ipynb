# STARTING Matplotlib

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

Changing style to look similar to R ggplot2 package

```python
plt.style.use('ggplot')
```

VARIABLE DECLARATION (they are data arrays, and functions on that data)

```python
x: np.ndarray = np.linspace(0, 20, 1000) # 1000 evenly spaced values from 0 to 20
x2 = np.linspace(0, 20, 100)
print(type(x))
print(type(x2)) # Type is recognized both we declare typed var or not
cos_x = np.cos(x)
sin_x2 = np.sin(x2)
tan_x = np.tan(x)
```

### OPTION 1: 
Use directly plt.plot(var1,fun(var1),var2,fun(var2),...) without create figure and axis.

```python
plt.plot(x2, x2**3, label='x**3')
plt.plot(x2, x2**2, marker='.', color='grey', label='x**2')
plt.legend() # This .legend() affects only plots above
```

### OPTION 2: 
CREATING FIGURE AND AXIS (basic elements for all plots)

```python
figure = plt.figure() # 'Empty' constructor, but we can add some parameters to customize its appearance
axis = plt.axes() # 'Empty', same as above
```

CUSTOMIZE PLOT APPEARANCE<br>
- axis.method() -> Acts on figure object plots<br>
- plt.method() -> Acts on last graph (polygonal if Bar and Box plots are comented)<br>
- .set(attrname = attrvalue, attrname2 = ...) or axis.set_attrname(attrvalue) or plt.attrname(attrvalue)

ASSIGN VARIABLE-FUNCTION TO AXIS created above<br>
**LINE PLOT**

```python
def restart_axis(title:str):
    axis.clear() # Restart both appearance&plots
    axis.set_title(title, color='b', fontsize=10)
    axis.set(xlim=[0, 22], ylim=[-3,3]) # To show convenient tan(x) (that has infinite limits)
    axis.margins(x=0.0, y=0.5) # x=0% y=50% of graphic rise
    axis.set_xlabel('x')
    axis.set_ylabel('y(x)')
```

```python
restart_axis('Trigonometric functions')
axis.plot(x, cos_x, linewidth=1.2, linestyle='--', color="yellow", label='cos(x)')
axis.plot(x, tan_x, color="pink", label='tan(x)')
axis.legend(fontsize=7) # labels for each function
figure
```

SAVING plot as image file

```python
figure.savefig('plotImages/trig_plot.png')
```

### PLOT TYPES

- **Line plot**

```python
restart_axis('Line plot')
axis.plot(x, x/10, color="pink", label='x/10')
axis.legend(fontsize=7)
figure.savefig('plotImages/line_plot.png')
figure
```

- **Scatter plot**

```python
restart_axis('Scatter plot')
x = np.random.random(30)*20
y = x * np.random.random(30)/10
axis.scatter(x, y, color="orange", marker='o', label='Random values')
axis.legend(fontsize=7)
figure.savefig('plotImages/scatter_plot.png')
figure
```

- **Histogram plot** *(We can built histograms directly from pandas DataFrame and Series, but to save or manipulate them is better working with axis and figure as above)*

```python
# Histogram from pandas DataFrame
data_frame = pd.DataFrame(np.random.random(30)*20)
plot1 = data_frame.hist(bins=20, color="orange", label='Random values')
plot2 = data_frame.plot(kind='hist', bins=20, color="orange", label='Random values')
print(type(plot1)) # numpy ndarray
print(type(plot2)) # matplotlib AxesSubplot
#plot2.legend(fontsize=7) # Doesn't work
#plot2.savefig('histogram_plot.png') # Doesn't work
```

```python
# Histogram with Figure and Axes
restart_axis('Histogram plot')
axis.set_ylim([0,10])
x = np.random.random(100)*20
axis.hist(x, bins=20, color="orange", label='Random values')
axis.legend(fontsize=7)
figure.savefig('plotImages/histogram_plot.png')
figure
```

- **Bars plot**

```python
restart_axis('Bars plot')
axis.set_ylim([0,3])
x = np.random.random(30)*20
y = x * np.random.random(30)/6
axis.bar(x, y, color="red", label='Random values')
axis.legend(fontsize=7)
figure.savefig('plotImages/bars_plot.png')
figure
```

- **Box plot** *(We can built box plots directly from pandas DataFrame and Series, but to save or manipulate them is better working with axis and figure as above)*

```python
# Boxplot from DataFrame
data_frame = pd.DataFrame(np.random.random(30)*20)
plot = data_frame.plot(kind='box', label='Random values')
print(type(plot)) # matplotlib AxesSubplot
#plot.legend(fontsize=7) # Doesn't work
#plot.savefig('histogram_plot.png') # Doesn't work
```

```python
# Boxplot with Figure and Axes
restart_axis('Box plot')
axis.set_xlim([-1,20])
axis.set_ylim([0.5,1.5]) # To show shapely
x = np.random.random(30)*20
axis.boxplot(x, vert=False)
figure.savefig('plotImages/box_plot.png')
figure
```
